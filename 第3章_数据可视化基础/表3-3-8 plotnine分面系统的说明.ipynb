```python
from plotnine import *
from plotnine.data import mpg


mpg
```

```python
t = (ggplot(mpg, aes('cty', 'hwy', fill='fl'))
     + geom_point(size=3, stroke=0.3, alpha=0.8, show_legend=False)
     + scale_fill_hue(s=0.90, l=0.65, h=0.0417, color_space='husl')
     )

t1 = (t + facet_grid('.~ fl')
      + theme_matplotlib()
      + theme(text=element_text(size=12, colour="black"),
              strip_text=element_text(size=15, colour="black"),
              strip_background=element_rect(color='k'),
              aspect_ratio=1.3,
              dpi=100,
              figure_size=(10, 10)))
print(t1)
# t1.save('分面1.pdf')
```

```python
t2 = (t + facet_grid('year ~ .')
      + theme_matplotlib()
      + theme(text=element_text(size=12, colour="black"),
              strip_text=element_text(size=15, colour="black"),
              strip_background=element_rect(color='k'),
              aspect_ratio=0.5,
              dpi=100,
              figure_size=(4, 4)))
print(t2)
# t2.save('分面2.pdf')
```

```python
t3 = (t + facet_grid('year ~ fl')
      + theme_matplotlib()
      + theme(text=element_text(size=12, colour="black"),
              strip_text=element_text(size=15, colour="black"),
              strip_background=element_rect(color='k'),
              aspect_ratio=1,
              dpi=100,
              figure_size=(9, 9)))
print(t3)
# t3.save('分面3.pdf')
```
